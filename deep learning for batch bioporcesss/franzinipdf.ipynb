```python
import PyPDF2
from PyPDF2 import PdfReader, PdfWriter

def crop_pdf(input_pdf_path, output_pdf_path, split_ratio=0.5):
    """
    Crop PDF pages into two halves (left and right) and join them into a new PDF.
    
    Args:
        input_pdf_path (str): Path to the input PDF file.
        output_pdf_path (str): Path to the output PDF file.
        split_ratio (float): Ratio to control where the split happens. 
                             A value between 0 and 1, where 0.5 is the exact middle.
    """
    # Open the input PDF file
    input_pdf = PdfReader(open(input_pdf_path, "rb"))
    output_pdf = PdfWriter()

    # Iterate through each page in the input PDF
    for page_num in range(len(input_pdf.pages)):
        page = input_pdf.pages[page_num]
        media_box = page.mediabox
        width = media_box.right
        height = media_box.top
        #print(width , split_ratio)

        # Adjust the split position based on split_ratio
        split_position = int(width) * int(split_ratio)

        # Create the left half
        left_page = PdfWriter()
        left_page.add_page(page)
        left_page.pages[0].mediabox.upper_right = (split_position, height)

        # Create the right half
        right_page = PdfWriter()
        right_page.add_page(page)
        right_page.pages[0].mediabox.upper_left = (split_position, 0)

        # Add the cropped pages to the output PDF
        output_pdf.add_page(left_page.pages[0])
        output_pdf.add_page(right_page.pages[0])
        print(f"Cropped page {page_num} into left and right halves.")

    # Write the output PDF to a file
    with open(output_pdf_path, "wb") as output_file:
        output_pdf.write(output_file)
```

```python
from pdf2image import convert_from_path
from PIL import ImageEnhance
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter
import os

def adjust_brightness(image, brightness_factor):
    """
    Adjust the brightness of an image.
    
    Args:
        image (PIL.Image): The image to adjust.
        brightness_factor (float): A factor to control brightness. 
                                   >1.0 increases brightness, <1.0 decreases.
    
    Returns:
        PIL.Image: Brightness-adjusted image.
    """
    enhancer = ImageEnhance.Brightness(image)
    return enhancer.enhance(brightness_factor)

def edit_pdf(input_pdf_path, output_pdf_path, brightness_factor=1.5):
    """
    Extract pages from a PDF, adjust the brightness, and reassemble them into a new PDF.
    
    Args:
        input_pdf_path (str): Path to the input PDF file.
        output_pdf_path (str): Path to save the output PDF.
        brightness_factor (float): Factor to adjust brightness (default 1.5 for 50% brighter).
    """
    # Convert PDF pages to images
    images = convert_from_path(input_pdf_path)
    
    # Create a new PDF file to save the adjusted pages
    output_pdf = canvas.Canvas(output_pdf_path, pagesize=letter)
    
    for i, image in enumerate(images):
        # Adjust brightness of the image
        bright_image = adjust_brightness(image, brightness_factor)
        
        # Save the image temporarily
        temp_image_path = f"temp_page_{i}.png"
        bright_image.save(temp_image_path)
        
        # Add the brightened image to the PDF
        output_pdf.drawImage(temp_image_path, 0, 0, width=letter[0], height=letter[1])
        output_pdf.showPage()  # Move to next page

        # Remove the temporary image
        os.remove(temp_image_path)

    output_pdf.save()
    print(f"Brightness adjusted PDF saved at {output_pdf_path}")
```

```python

# Example usage
input_pdf_path = r"D:\ALL CODES\REINCEFORCEMENT LEARNING\Papers\PAPERS\chemE RL papers\deep learning for batch bioporcesss\Fluid_mechanics_with_engineering_applications_Finnemore,_E_John;.pdf"
output_pdf_path = "hello.pdf"
edit_pdf(input_pdf_path, output_pdf_path, brightness_factor=1.5)  # Adjust brightness by 50%
```
